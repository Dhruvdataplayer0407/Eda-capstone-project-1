# **Data visulaization**

```python
# 1) total number of apps in each category
sns.set_style('darkgrid')
plt.figure(figsize=(10, 5))
sns.countplot(x='Category', data=df_dataset1)
plt.title('Number of Apps Per Category')
plt.xticks(rotation=90)
plt.ylabel('Number of Apps')
plt.show()
```

# **conclusion**
Here, we can see clearly that family category is on top position in terms of total number of apps,followed by game,persolization category
and Beauty category is on bottom postion in terms of total number of apps

```python
# 2) show the distribution of rating
plt.figure(figsize=(10, 5))
sns.countplot(x='Rating', data=df_dataset1)
plt.title('Rating Distribution')
plt.xticks(rotation=90)
plt.ylabel('Number of Apps')
plt.show()
```

# **conclusion**
As we can see clearly from above bar graph that approx more than 1000 apps have a 4.5 rating

```python
# 3) most freqent rating that are used in apps
plt.figure(figsize=(15,5))
plt.xlabel("Rating")
plt.ylabel("Frequency")
graph = sns.kdeplot(df_dataset1.Rating, color="#4B0751", shade = True)
plt.title('Most Frequent Rating',size = 20);
```

# # 4) ***Content Rating***

```python
plt.figure(figsize=(15,10))
pd.value_counts(df_dataset1["Content Rating"]).plot(kind = 'pie',  autopct='%2.4f%%', shadow=True)
plt.show()
```

# **Conclusion**
from above pie cart we can say clearly that maximum percentage of content ,installed on playstore is for Everyone,followed by Teen Mature17+ ,etc

# #5)  **Content Rating in free and Paid** 

```python
plt.figure(figsize=(15,10))
ax = sns.countplot(df_dataset1['Content Rating'],hue=df_dataset1['Type']);
plt.title("ContentRating in Free and Paid",size=20)

for p in ax.patches:
    ax.annotate(p.get_height(),(p.get_x() + p.get_width()/2,p.get_height()),ha='center',size= 12)
    
plt.show()
```

# **conclusion**
with the help of above graph we can say that in the context  Content Rating ,tf everyone has a maximum number of free apps on playstore .
